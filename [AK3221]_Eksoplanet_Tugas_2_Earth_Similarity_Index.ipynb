##Perhitungan ESI Untuk Eksoplanet

###Insolation Flux

untuk menghitung fluks insolasi menggunakan persamaan $$ \frac{S}{S_\oplus} = \bigg(\frac{L_*}{L_⊙}\bigg)\bigg(\frac{AU}{a}\bigg)^2$$
dengan $S$ adalah insolasi (daya per unit area dalam unit Bumi), $a$ merupakan sumbu semimayor. Sumber : https://exoplanetarchive.ipac.caltech.edu/docs/poet_calculations.html

```python
#Insolation Flux (S)
S = 10**df['st_lum']*(1/df['pl_orbsmax'])**2
S2 = df2['st_lum']*(1/df2['a'])**2
S
```

```python
df['st_flu'] = S
df2['st_flu2'] = S2
```

###Perhitungan ESI

A simple ESI expression, as a function of a planet's stellar flux, and radius or mass, is used for exoplanets since there is little information about them. The ESI is the given by a distance metric:

$$ESI(S,R) = 1 - \sqrt{\frac{1}{2} \bigg[ \bigg( \frac{S-S_\oplus}{S+S_\oplus}\bigg)^2 + \bigg( \frac{R-R_\oplus}{R+R_\oplus}\bigg)^2\bigg]}$$

where S is stellar flux, R is radius, $S_⊕$ is Earth's solar flux, and $R_⊕$ is Earth's radius. This ESI expression uses the quadratic mean as the distance metric, which is very convenient to interpret statistically (e.g., as a chi-squared distribution). This expression can be used for those planets detected by the transit method (e.g., Kepler, K2, TESS, Plato, etc.) where only the planet's radius is known. Sumber : https://phl.upr.edu/projects/earth-similarity-index-esi

```python
df['a']  = ((df['st_flu'] - 1)/(df['st_flu'] + 1))**2
df['b']  = ((df['pl_rade']- 1)/(df['pl_rade']+ 1))**2

df2['c'] = ((df2['st_flu2'] - 1)/(df2['st_flu2'] + 1))**2
df2['d'] = ((df2['pl_rade']- 1)/(df2['pl_rade']+ 1))**2
```

```python
df2['esi2'] = 1 - np.sqrt(0.5*(df2['c'] + df2['d']))
max(df2['esi2'])
```

```python
df['esi'] = 1 - np.sqrt(0.5*(df['a'] + df['b']))
max(df['esi'])
```
